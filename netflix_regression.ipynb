```python
import pandas as pd
B = pd.read_csv("net.csv")
B.head()
```

```python
import seaborn as sns
sns.lineplot(x = "Date",y = "Adj Close",data = B)
```

```python
B.dropna(inplace = True)
```

```python
from scipy import stats
import numpy as np
z = np.abs(stats.zscore(B[["Open","High","Low","Close","Adj Close"]]))
print(z)
```

```python
z.max()
```

```python
threshold_z = 2
index = np.where(z>threshold_z)[0]
no_outliers = B.drop(index)
print(no_outliers.shape)
```

```python
C = no_outliers
```

```python
X = C.drop(columns = ["Date","Close"])
Y = C["Close"]
```

```python
from sklearn.model_selection import train_test_split,GridSearchCV
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.3)
from sklearn.ensemble import AdaBoostRegressor
A = AdaBoostRegressor()
params_grid = {"n_estimators":[100,200,400,500],"learning_rate":[0.01,0.1,0.5,1]}
grid = GridSearchCV(A,params_grid,cv = 10)
grid.fit(X_train,Y_train)
Y_predict = grid.predict(X_test)
print(grid.best_score_)
print(grid.best_params_)
```
